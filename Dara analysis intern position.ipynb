```python
def custom_contains(fullstring, substring):
    """
    checks to see if fullstring contains substring
    Input:
        fullstring: the full string
        substring: substring
    Output:
        Boolean value
    """
    if pd.isnull(fullstring):
        return False
    fullstring = fullstring.replace("'","")
    if substring in fullstring:
        return True
    else:
        return False
```

```python
custom_contains(df.loc[2,'Qualification'], "bachelors degree")
```
